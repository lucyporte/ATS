#Plotting the data

Standard approach

```python
import matplotlib.pyplot as plt
plt.plot(df.index.tolist(),df["Open"].tolist(),label="Amazon stock open")
```

```python
plt.plot(df["Open"])
```

Plotting multiple lines on same graph

```python
plt.plot(df.index.tolist(),df["Open"].tolist(),label="Open")
plt.plot(df.index.tolist(),df["Close"].tolist(),label="Close")
plt.legend()
```

Simpler method, just input the dataframe into the plot function

```python
plt.plot(df["Open"],label="Open")
plt.plot(df["Close"],label="Close")
plt.legend()
plt.title("Graph of Amazon stock price")
```

```python
df.plot(y=["Open","Close"])
```

The "best" way to plot data. Simply use the pandas plot function. For more info see the link below

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html

```python
df.plot(y=["Open", "Close"],title="Amazon stock price")
plt.show()
```
